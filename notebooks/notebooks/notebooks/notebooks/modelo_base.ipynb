# AN√ÅLISIS DE PATRONES - CONSUMO COMBUSTIBLE
## Modelo Base: An√°lisis Explicativo

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv('../data/datasets_anonimizados.csv')
print("‚úÖ Datos cargados")
df.head()
```

```python
# 1. C√°lculo eficiencia
df['eficiencia'] = df['combustible_principal_litros'] / df['kilometros_recorridos']

# 2. Meses m√°s y menos eficientes
mes_mejor = df['eficiencia'].idxmin()
mes_peor = df['eficiencia'].idxmax()
meses = ['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']

print("üéØ EFICIENCIA POR MES (L/km):")
for i, row in df.iterrows():
    print(f"   {meses[i]}: {row['eficiencia']:.4f}")

print(f"\nüìä RESUMEN:")
print(f"   ‚Ä¢ Mejor: {meses[mes_mejor]} ({df['eficiencia'].min():.4f} L/km)")
print(f"   ‚Ä¢ Peor:  {meses[mes_peor]} ({df['eficiencia'].max():.4f} L/km)")
print(f"   ‚Ä¢ Promedio: {df['eficiencia'].mean():.4f} L/km")
```

```python
# 3. Gr√°fico simple
plt.figure(figsize=(10, 4))
plt.plot(meses, df['eficiencia'], marker='o', linewidth=2)
plt.axhline(y=df['eficiencia'].mean(), color='red', linestyle='--', label='Promedio')
plt.title('EVOLUCI√ìN EFICIENCIA (L/km)')
plt.ylabel('Litros por Km')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("‚úÖ An√°lisis completado")
```
